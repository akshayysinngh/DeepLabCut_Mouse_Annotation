## Removing the unwanted rows

```python
df = df.drop([0,1,2], axis=0)
```

## Creating a dataframe for Mouse 1 

```python
mouse1_df = df.iloc[:,:18]
```

```python
mouse1_df
```

## Creating a dataframe for Mouse 2

```python
mouse2_df = df.iloc[:,18:]
```

```python
mouse2_df
```

```python
mouse2_df.columns = ['mouse1_topleft_x', 'mouse1_topleft_y', 'mouse1_rightdown_x',
       'mouse1_rightdown_y', 'mouse1_nose_x', 'mouse1_nose_y',
       'mouse1_leftear_x', 'mouse1_leftear_y', 'mouse1_rightear_x',
       'mouse1_rightear_y', 'mouse1_leftHip_x', 'mouse1_leftHip_y',
       'mouse1_rightHip_x', 'mouse1_rightHip_y', 'mouse1_tailbase_x',
       'mouse1_tailbase_y', 'mouse1_tailend_x', 'mouse1_tailend_y']
```

## Combining both dataframe in one

```python
concatenated_mouse_data = pd.concat([mouse1_df, mouse2_df], axis = 0 )
```

```python
concatenated_mouse_data
```

```python
concatenated_mouse_data = concatenated_mouse_data.dropna()
```

```python
concatenated_mouse_data.columns = ['mouse_topleft_x', 'mouse_topleft_y', 'mouse_rightdown_x',
       'mouse_rightdown_y', 'mouse_nose_x', 'mouse_nose_y',
       'mouse_leftear_x', 'mouse_leftear_y', 'mouse_rightear_x',
       'mouse_rightear_y', 'mouse_leftHip_x', 'mouse_leftHip_y',
       'mouse_rightHip_x', 'mouse_rightHip_y', 'mouse_tailbase_x',
       'mouse_tailbase_y', 'mouse_tailend_x', 'mouse_tailend_y']
```

```python
concatenated_mouse_data.dtypes
```

## Converting Object to float 

```python
concatenated_mouse_data = concatenated_mouse_data.astype(float)
```

```python
concatenated_mouse_data.dtypes
```

```python
concatenated_mouse_data["avg_cord_x"] = (concatenated_mouse_data["mouse_topleft_x"] + concatenated_mouse_data["mouse_rightdown_x"]) / 2
concatenated_mouse_data["avg_cord_y"] = (concatenated_mouse_data["mouse_topleft_y"] + concatenated_mouse_data["mouse_rightdown_y"]) / 2
concatenated_mouse_data
```

## Plot of mouse movement

```python
concatenated_mouse_data.plot(x = 'avg_cord_x', y = 'avg_cord_y')
```

## Test-Train Split

```python
from sklearn.model_selection import train_test_split
```

```python
X  = concatenated_mouse_data.iloc[: , 4:18]
Y  = concatenated_mouse_data.iloc[: , 18:]
X_train, X_test, Y_train, Y_test = train_test_split(X, Y,random_state=100, test_size=0.30)
```

```python
X_test.shape
```

```python
Y_train.shape
```

```python
Y_test.shape
```

## Building neural network

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
neural_model = Sequential()
neural_model.add(Dense(14, input_dim = 14, activation='relu'))
neural_model.add(Dense(14, activation = 'relu'))
neural_model.add(Dense(2))
```

""""

```python
with open('neural_model.json','r') as f:
    json = f.read()
neural_model = tf.keras.models.model_from_json(json)

neural_model.load_weights("ml_mouse_location.h5")
```

```python
neural_model.summary()
```

```python
neural_model.compile(loss='mean_squared_error',optimizer='adam', metrics=['accuracy'])
```

```python
neural_model.fit(X_train, Y_train, epochs = 100, validation_data = (X_test, Y_test))
```

```python
accuracy = neural_model.evaluate(X_test, Y_test, verbose=0)
print(accuracy[1]*100)
```

```python
history = neural_model.fit(X_train, Y_train, epochs=100, batch_size=1, validation_data=(X_test, Y_test))
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'validation'], loc='upper left')
plt.show()
```
